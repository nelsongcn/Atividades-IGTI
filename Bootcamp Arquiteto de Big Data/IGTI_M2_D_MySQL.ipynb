## Substituindo os dados nulos

```python
dados[dados.num_vitorias.isna()]
```

```python
dados = dados.fillna({'num_vitorias' : dados.total_partidas - dados.num_derrotas, 'num_derrotas' : dados.total_partidas - dados.num_vitorias,
             'total_partidas': dados.num_vitorias + dados.num_derrotas})
```

## Alterando o formato da data para se adequar ao MySQL

```python
dados.data_nascimento
```

```python
for data in dados.data_nascimento:
    data_sr = pd.to_datetime(data)
    change_format = data_sr.strftime('%Y-%m-%d')
    dados.replace(data, change_format, inplace = True)


    
```

## Exportando Para o MySQL

```python
import MySQLdb # importação da biblioteca mysqlclient
import sqlalchemy
```

```python
user = 'root'
password = '123456'
host = 'localhost'
database = 'desafiomii_v2' #schema

string_conexao = f'mysql://{user}:{password}@{host}/{database}'
string_conexao #testar a string de conexão com o banco de dados. 
```

```python
engine = sqlalchemy.create_engine(string_conexao)
conn = engine.connect()
```

```python
for dado in dados.itertuples():
    jogador = dado.jogador
    genero = dado.genero    
    data_nascimento = str(dado.data_nascimento)
    jogo = dado.jogo
    pais = dado.pais
    num_vitorias = dado.num_vitorias
    num_derrotas = dado.num_derrotas
    total_partidas = dado.total_partidas
    try:
        query = f"""insert into jogador (nome_jogador, genero, data_nascimento, nome_jogo, nome_pais, num_vitorias,
                    num_derrotas,  total_partidas ) 
                     values ('{jogador}', '{genero}', '{data_nascimento}', '{jogo}', '{pais}',
                     '{num_vitorias}', '{num_derrotas}', '{total_partidas}')"""      
        conn.execute(query)      
        print(f"Registro inserido com sucesso: {jogador}")
    except Exception as e:        
        print(f"Não foi possível inserir o registro {jogador}. O erro foi encontrado foi: {e}")   
```

```python
dados.to_csv('dados_jogadores-1_tratado.csv', sep = ';', encoding='latin-1', index=False)
```
